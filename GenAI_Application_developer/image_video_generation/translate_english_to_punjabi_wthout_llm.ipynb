```python
from deep_translator import GoogleTranslator

english_text = """
Hello everyone.
Welcome to the world of Artificial Intelligence.
I hope you enjoy learning Python.
"""

punjabi_text = GoogleTranslator(
    source='en',
    target='pa'
).translate(english_text)

print("Punjabi Translation:\n")
print(punjabi_text)
```
